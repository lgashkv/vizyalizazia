```python
max_GDP_per_capita = df[['Country or region', 'GDP per capita']].sort_values(by='GDP per capita', ascending=False).head(1)
max_Social_support = df[['Country or region', 'Social support']].sort_values(by='Social support', ascending=False).head(1)
max_Hle = df[['Country or region', 'Healthy life expectancy']].sort_values(by='Healthy life expectancy', ascending=False).head(1)
max_Ftmlc = df[['Country or region', 'Freedom to make life choices']].sort_values(by='Freedom to make life choices', ascending=False).head(1)
max_Generosity = df[['Country or region', 'Generosity']].sort_values(by='Generosity', ascending=False).head(1)
max_Poc = df[['Country or region', 'Perceptions of corruption']].sort_values(by='Perceptions of corruption', ascending=False).head(1)
max_ozenka = pd.concat([max_GDP_per_capita ,max_Social_support,max_Hle,max_Ftmlc,max_Generosity,max_Poc])
max_ozenka.set_index('Country or region', inplace=True)
```

```python
ax = max_ozenka.plot(kind='bar'),plt.xlabel('Максимальная оценка'),plt.title('Максимальная оценка по градациям')
plt.xticks(rotation=60, horizontalalignment='right',fontsize=10)
ax
Qatar = df.loc[df['Country or region'] =='Qatar'] 
Iceland = df.loc[df['Country or region'] =='Iceland'] 
Singapore = df.loc[df['Country or region'] =='Singapore'] 
Uzbekistan = df.loc[df['Country or region'] =='Uzbekistan'] 
Myanmar = df.loc[df['Country or region'] =='Myanmar'] 
top = pd.concat([Qatar ,Iceland,Singapore,Uzbekistan,Myanmar])
dftop = top[['Country or region', 'Score','GDP per capita', 'Social support','Healthy life expectancy', 'Freedom to make life choices','Generosity','Perceptions of corruption']]
dftop.set_index('Country or region', inplace=True)
ax = dftop.plot(kind='barh', stacked=True),plt.title('Остальные оценки стран набравших максимальные очки')
plt.xticks(rotation=60, horizontalalignment='right',fontsize=12)
ax
```

```python
# В Катаре лучше всего ВВП, что не отменяет коррупцию и мешает людям чувствовать себя достаточно свободно. 
#В Сингапуре взаимосвязь самые низкие показатели по коррупции и самые лучшие показатели по медицине
# Исландия -  самая высокая оценка по соц.поддержке! Входит в топ стран по уровню счастья. Остальные показатели сравнимы.
#Узбекистан - выше всех оценил свои возможности по изменению жизни, является самым оптимистичным. 
```
